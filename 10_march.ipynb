Q2: Python Function to Estimate Population Mean

```python
def estimate_population_mean(sample_mean, sample_std, sample_size):
    import math
    standard_error = sample_std / math.sqrt(sample_size)
    return sample_mean, standard_error


sample_mean = 50
sample_std = 10
sample_size = 30

mean, std_error = estimate_population_mean(sample_mean, sample_std, sample_size)
print(f"Sample Mean: {mean}, Standard Error: {std_error}")
```

### Q5: Python Script for Hypothesis Test on Difference Between Two Population Means

```python
import scipy.stats as stats

def two_sample_t_test(sample1, sample2):
    t_stat, p_value = stats.ttest_ind(sample1, sample2)
    return t_stat, p_value

# Example usage:
sample1 = [85, 90, 78, 92, 88]  # Sample data for population 1
sample2 = [75, 80, 72, 78, 74]  # Sample data for population 2

t_stat, p_value = two_sample_t_test(sample1, sample2)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

## Q10: Python Program for Two-Sample T-Test for Independent Samples

```python
import numpy as np
import scipy.stats as stats

def two_sample_t_test(sample1, sample2):
    t_stat, p_value = stats.ttest_ind(sample1, sample2)
    return t_stat, p_value

# Example usage:
np.random.seed(0)
sample1 = np.random.normal(100, 10, 30)  # Sample 1
sample2 = np.random.normal(102, 10, 30)  # Sample 2

t_stat, p_value = two_sample_t_test(sample1, sample2)

print(f"T-statistic: {t_stat}, P-value: {p_value}")
```
